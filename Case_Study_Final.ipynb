### **OUTLIER DETECTION**

**INFERENCE:**

An outlier is an object that deviates significantly from the rest of the objects. They can be caused by measurement or execution error.
So, we are using boxplot to detect its deviation.


```python
plt.figure(figsize=(16,8))
plt.subplot(2,3,1)
sns.boxplot(df['Confirmed'], color='#F07B3F')
plt.subplot(2,3,2)
sns.boxplot(df['Recovered'], color='#00ADB5')
plt.subplot(2,3,3)
sns.boxplot(df['Deceased'], color='#E84545')
```

**FINDING INTERQUARTILE RANGE , UPPER AND LOWER LIMIT**

```python
p_25 = df['Confirmed'].quantile(0.25)
p_75 = df['Confirmed'].quantile(0.75)

p1_25 = df['Recovered'].quantile(0.25)
p1_75 = df['Recovered'].quantile(0.75)

p2_25 = df['Deceased'].quantile(0.25)
p2_75 = df['Deceased'].quantile(0.75)

iqr=p_75-p_25
iqr1=p1_75-p1_25
iqr2=p2_75-p2_25
```

**FINDING UPPER AND LOWER RANGE**

```python
ul = p_75 + 1.5 * iqr
ll = p_25 - 1.5 * iqr

ul_1 = p1_75 + 1.5 * iqr1
ll_1 = p1_25 - 1.5 * iqr1

ul_2 = p2_75 + 1.5 * iqr2
ll_2 = p2_25 - 1.5 * iqr2
```

**MANUAL DETECTION OF OUTLIERS**

```python
df[(df['Confirmed'] > ul) | (df['Confirmed'] < ll)]
```

```python
df[(df['Recovered'] > ul_1) | (df['Recovered'] < ll_1)]
```

```python
df[(df['Deceased'] > ul_2) | (df['Deceased'] < ll_2)]
```

**CAPPING AND TRIMMING**

*i.e removal of outliers*

```python
n_df = df.copy()
n_df['Confirmed'] = np.where(
    n_df['Confirmed'] > ul,
    ul,
    np.where(
        n_df['Confirmed'] < ll,
        ll,
        n_df['Confirmed']
    )
)

n_df['Recovered'] = np.where(
    n_df['Recovered'] > ul_1,
    ul_1,
    np.where(
        n_df['Recovered'] < ll_1,
        ll_1,
        n_df['Recovered']
    )
)

n_df['Deceased'] = np.where(
    n_df['Deceased'] > ul_2,
    ul_2,
    np.where(
        n_df['Deceased'] < ll_2,
        ll_2,
        n_df['Deceased']
    )
)
```

# **HYPOTHESIS TESTING**

```python
import numpy as np
import scipy.stats as st
```

```python
ht = n_df.copy()
```

```python
print('Mean               : ', ht['Deceased'].mean())
print('Standard Deviation : ', np.std(ht['Deceased']))
print('Variance           : ', np.var(ht['Deceased']))
```

```python
print('H0 : mu = ',ht['Deceased'].mean())
print('H1 : mu â‰  ',ht['Deceased'].mean())
```

### **Z - TEST**

```python
#H0 : Î¼ = 33.11911487758945, H1 :  Î¼ â‰  33.11911487758945
n = 50
xbar = 43
mu = 33.11911487758945
sigma = 36.58095151980261
alpha = 0.05
```

```python
from scipy import stats
from statsmodels.stats import weightstats as stests
ztest ,pval = stests.ztest(ht['Deceased'], x2=None, value=43)
print(float(pval))
if pval<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

```python
z_critical = abs(st.norm.ppf(0.05))
z_critical
```

```python
z = (xbar-mu) / (sigma/np.sqrt(n))
z
```

```python
if(z < z_critical):
  print('Accept Null Hypothesis')
else:
  print('Reject Null Hypothesis')
```

### **T - TEST**

```python
mu=33

from scipy import stats as st

t_value,p_value=stats.ttest_1samp(ht['Deceased'],mu)

one_tailed_p_value=float("{:.6f}".format(p_value/2)) # Since alternative hypothesis is one tailed, We need to divide the p value by 2.

print('Test statistic is %f'%float("{:.6f}".format(t_value)))

print('p-value for one tailed test is %f'%one_tailed_p_value)

alpha = 0.05

if one_tailed_p_value<=alpha:

    print('Conclusion','n','Since p-value(=%f)'%p_value,'<','alpha(=%.2f)'%alpha,'''We reject the null hypothesis H0.''')

else:

    print('Conclusion','n','Since p-value(=%f)'%one_tailed_p_value,'>','alpha(=%.2f)'%alpha,'''We do not reject the null hypothesis H0.''')
```

### **CONFIDENCE INTERVALS**

```python
# 95% confidence interval for population mean weight
st.norm.interval(alpha=0.95,loc=np.mean(ht['Deceased']),scale=st.sem(ht['Deceased']))
```

```python
#99% confidence interval for same sample
st.norm.interval(alpha=0.99,loc=np.mean(ht['Deceased']),scale=st.sem(ht['Deceased']))
```

```python
# Initialzing empty lists to append all model's name and corresponding name
acc = []
model = []
err = []
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_predict
```

```python
def crossValEvaluate(model, X, y):
  y_pred = cross_val_predict(model, X, y, cv=3);

  scores = mean_squared_error(y_pred, y)
  print(f'Train Error: {np.sqrt(scores)}')
```

```python
def testSetEvaluate(model, X_train, y_train, X_test, y_test):
  model.fit(X_train, y_train)
  y_pred = model.predict(X_test)

  scores = mean_squared_error(y_pred, y_test)
  print(f'Test Error: {np.sqrt(scores)}')

  return scores
```
